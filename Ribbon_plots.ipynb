To truly understand why Plotly stands out, it helps to look at the philosophy behind Matplotlib and Seaborn, which represent the "traditional" way of plotting data in Python.

Here is how they differ across the core areas of data visualization: **bold text**

1. Interactivity vs. Static Images
Matplotlib & Seaborn: Generate static, flat images (like a PNG or JPEG). Once the chart is drawn, you cannot interact with it. If data points are crowded together, they overlap, and you cannot zoom in to separate them.

Plotly: Generates dynamic, HTML-based charts. You can hover your mouse over any data point to see its exact values, click-and-drag to zoom into crowded areas, and click legends to turn specific data lines on or off dynamically.

2. The Core Technology Under the Hood
Matplotlib: The grandfather of Python plotting (created in 2003). It was designed to mimic the plotting capabilities of MATLAB. It draws shapes onto a canvas pixel by pixel.

Seaborn: Built on top of Matplotlib. It takes Matplotlib's powerful engine but applies beautiful default themes and simplifies the code needed to create complex statistical plots (like heatmaps or violin plots).

Plotly: Built on JavaScript (Plotly.js). When you run Plotly code in Python, it actually translates your data into JSON format and uses web browser technology to render the graphics.

3. Use Cases: When to Use Which?
Use Matplotlib / Seaborn if:
You are printing a physical report, writing an academic paper, or exporting a static PDF.

You are doing rapid, basic data exploration in your notebook where you just need to glance at a quick histogram.

You need pixel-perfect control over a static image layout.

Use Plotly if:
You are building a dashboard or a web application.

Your data has multiple dimensions where hovering over a point provides crucial context.

You want to share an HTML file with a client or teammate so they can explore the data themselves without needing to read your code.

You need complex, flowing visualizations like the Sankey Diagrams or Sunburst Charts we built earlier (which are incredibly tedious or impossible to build beautifully in Matplotlib).

````python
# Step 1: Install plotly if it's not already available (Colab usually has it pre-installed)
!pip install plotly -q

import plotly.graph_objects as go
import pandas as pd

# Step 2: Define your source categories (Left side) and target categories (Right side)
# We will create a unique list of nodes.
# Index 0-11 are the Sectors, Index 12-14 are the Recipients.
nodes = [
    # Left side nodes (Sectors)
    "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics",
    "Transportation", "Construction", "Finance/Insur/RealEst", "Health",
    "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
    # Right side nodes (Recipients)
    "Democrats", "Republicans", "Outside Spending Grps"
]

# Step 3: Define the flows (source index -> target index -> value/amount)
# Node indices map directly to the 'nodes' list above (e.g., 0 = Lawyers & Lobbyists, 12 = Democrats)
data = [
    # Source, Target, Value
    [0, 12, 50], [0, 13, 20], [0, 14, 5],   # Lawyers & Lobbyists
    [1, 12, 45], [1, 13, 25], [1, 14, 10],  # Other
    [2, 12, 15], [2, 13, 35], [2, 14, 8],   # Defense
    [3, 12, 30], [3, 13, 15], [3, 14, 12],  # Communic/Electronics
    [4, 12, 12], [4, 13, 22], [4, 14, 4],   # Transportation
    [5, 12, 10], [5, 13, 28], [5, 14, 6],   # Construction
    [6, 12, 60], [6, 13, 75], [6, 14, 30],  # Finance/Insur/RealEst
    [7, 12, 40], [7, 13, 30], [7, 14, 5],   # Health
    [8, 12, 55], [8, 13, 40], [8, 14, 80],  # Ideology/Single-Issue
    [9, 12, 8],  [9, 13, 24], [9, 14, 3],   # Agribusiness
    [10, 12, 20], [10, 13, 30], [10, 14, 10],# Misc Business
    [11, 12, 12], [11, 13, 48], [11, 14, 15] # Energy/Nat Resources
]

# Convert the data list into a cleaner DataFrame structure
df = pd.DataFrame(data, columns=['source', 'target', 'value'])

# Step 4: Define custom colors for the target nodes to match your image
# Democrats = Blue, Republicans = Red, Outside Spending = Yellow/Gold
# Left nodes can remain a neutral grey or inherit transparency
node_colors = ["#7f7f7f"] * 12 + ["#3366cc", "#ff6666", "#ffcc00"]

# Assign ribbon colors based on their target (where they flow to)
def get_link_color(target_index):
    if target_index == 12:   return "rgba(51, 102, 204, 0.4)"   # Translucent Blue
    elif target_index == 13: return "rgba(255, 102, 102, 0.4)"  # Translucent Red
    else:                    return "rgba(255, 204, 0, 0.4)"    # Translucent Yellow

df['link_color'] = df['target'].apply(get_link_color)

# Step 5: Construct the Sankey Diagram
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = nodes,
      color = node_colors
    ),
    link = dict(
      source = df['source'].tolist(),
      target = df['target'].tolist(),
      value = df['value'].tolist(),
      color = df['link_color'].tolist()
  ))]) # Removed the extra ']' here

# Step 6: Customize layout text, font, and dimensions
fig.update_layout(
    title_text="Contributions During 2018 Election Cycle<br><sub>(by Sector)</sub>",
    title_x=0.5, # Center the title
    font_size=12,
    width=700,
    height=800,
    plot_bgcolor='white',
    paper_bgcolor='white'
)

# Step 7: Display the interactive graph inside Google Colab
fig.show()
````

````python
import plotly.graph_objects as go

fig = go.Figure(go.Sankey(
    arrangement="snap",
    node=dict(
        pad=20,
        thickness=25,
        line=dict(color="black", width=0.5),
        label=nodes,
        color=node_colors
    ),
    link=dict(
        source=df['source'],
        target=df['target'],
        value=df['value'],
        color=df['link_color']
    )
))

fig.update_layout(
    title="Election Contributions - Compact Flow View",
    font_size=12,
    height=700,
    width=900
)

fig.show()
````

````python
import pandas as pd
import plotly.express as px

df_bar = df.copy()

# map target names
df_bar["target_name"] = df_bar["target"].map({
    12: "Democrats",
    13: "Republicans",
    14: "Outside"
})

df_bar["source_name"] = df_bar["source"].map(dict(enumerate(nodes)))

fig = px.bar(
    df_bar,
    x="source_name",
    y="value",
    color="target_name",
    title="Contributions by Sector (Stacked View)",
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
````

````python
import plotly.express as px

df_sun = df.copy()

df_sun["source_name"] = df_sun["source"].map(dict(enumerate(nodes)))
df_sun["target_name"] = df_sun["target"].map({
    12: "Democrats",
    13: "Republicans",
    14: "Outside Spending"
})

fig = px.sunburst(
    df_sun,
    path=["source_name", "target_name"],
    values="value",
    title="Contribution Breakdown (Hierarchy View)"
)

fig.show()
````

````python
import plotly.express as px

fig = px.treemap(
    df_sun,
    path=["target_name", "source_name"],
    values="value",
    title="Treemap of Contributions"
)

fig.show()
````

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

for _, row in df.iterrows():
    G.add_edge(nodes[row["source"]], nodes[row["target"]], weight=row["value"])

plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", arrows=True)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Network Graph of Contributions")
plt.show()
````

````python
# Step 1: Install plotly if needed
!pip install plotly -q

import plotly.express as px
import pandas as pd

# Step 2: Re-create the complete flow dataset from the first image
# Mapping every sector contribution to its respective recipient group
data = {
    "Total": ["2018 Election Cycle"] * 36,
    "Recipient": [
        "Democrats", "Democrats", "Democrats", "Democrats", "Democrats", "Democrats",
        "Democrats", "Democrats", "Democrats", "Democrats", "Democrats", "Democrats",
        "Republicans", "Republicans", "Republicans", "Republicans", "Republicans", "Republicans",
        "Republicans", "Republicans", "Republicans", "Republicans", "Republicans", "Republicans",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps"
    ],
    "Sector": [
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources"
    ],
    "Amount (Millions)": [
        50, 45, 15, 30, 12, 10, 60, 40, 55, 8, 20, 12,  # Democrats
        20, 25, 35, 15, 22, 28, 75, 30, 40, 24, 30, 48,  # Republicans
        5, 10, 8, 12, 4, 6, 30, 5, 80, 3, 10, 15        # Outside Spending
    ]
}

df = pd.DataFrame(data)

# Step 3: Set up specific color mappings to resemble the original color choices
color_map = {
    "2018 Election Cycle": "#d3d3d3",       # Grey root
    "Democrats": "#3366cc",                 # Blue
    "Republicans": "#ff6666",               # Red
    "Outside Spending Grps": "#ffcc00"      # Yellow/Gold
}

# Step 4: Create the Sunburst Chart
fig = px.sunburst(
    df,
    path=["Total", "Recipient", "Sector"],  # Moving from the inside ring outwards
    values="Amount (Millions)",
    color="Recipient",                      # Base sector colors on the recipient node
    color_discrete_map=color_map,
    title="Contributions During 2018 Election Cycle (Hierarchical Structure)"
)

# Step 5: Clean up layout text styling and sizing
fig.update_layout(
    title_x=0.5,
    width=800,
    height=800,
    font_family="Arial"
)

# Optional: Adjust trace to display both labels and values cleanly inside the rings
fig.update_traces(
    textinfo="label+value"
)

# Step 6: Render chart inside Google Colab
fig.show()
````
